# SLICING

Slicing es una forma de “rebanar” (extraer partes) de una secuencia en Python, como listas, strings, tuplas o arrays de NumPy.

La sintaxis general es:

secuencia[inicio:fin:paso]


inicio → índice desde donde empieza (incluido).

fin → índice donde termina (excluido).

paso → cada cuántos elementos tomar.

```python
palabra_2 = "PYTHON"
v = palabra_2[0:-4]
print(v)
print(palabra_2[3:-2])

cadenaa = "Hola"
# cadenaa[4]  #  string index out of range

cadenaa[:99]  # con el scling el rango no me da error



########################3

# Inmutabilidad
# Una propiedad de las cadenas es que no se pueden modificar. Si intentamos reasignar un carácter, no nos dejará

al = "Juan"
# al[0] = "j"  # 'str' object does not support item assignment

al = "N" + al[0:1]
al
```
