# LISTAS
Una lista en Python es una especie de "caja" o "bolsa" donde puedo guardar varios elementos juntos, sin importar si son números, palabras u otras cosas, y además los puedo ordenar, cambiar, agregar o quitar cuando quiera.

```python
# Podemos imaginar una lista en py, como una lista de compras
# Las listas en py se define con [] corchetes, y los valores se separan por comas

"""
Puedes agregar más cosas: "Harina".

Puedes cambiar "pan" por "arepas".

Puedes quitar algo que ya no necesitas.

Y además la lista mantiene el orden en que pusiste las c

las listas a diferencia de los carchetes, son mutables, es decir que puedes cambiar su contenido sin problema.
"""
compras = ["arroz", "Leche", "Huevos", "Pan"]
print(compras)

# tambien se puede crear una lista vacia
vacia = []
vacia.append("Azucar")
print(vacia)

# Una lista es una colección ordenada de elementos que puedo modificar, 
# como una lista de mercado donde agrego, cambio o borro cosas según lo necesite


# En una lista puedo tener diferentes tipos de datos
mi_lista = [23, "Hola", 3.14, True]
print(mi_lista)

en = mi_lista[2]
print(en) 
```

```python
datos = [4,"Una cadena",-15,3.14,"Otra cadena"]
print(datos[2])
print(datos[-2:])

# Suma de listas
lista_1 = [1,2,3] # esta lista tiene los valores del 1 al 3 del índice del 0 al 2
lista_2 = [4,5,6] # esta lista tiene los valores del 4 al 6 del índice del 0 al 2
lista_3 = lista_1 + lista_2    # Creando una nueva lista con la suma de las dos anteriores
print(lista_3)

# son mutables

pares = [0,2,4,5,8,10]
print(pares)

pares[3] = 6
print(pares)

pares[5] = 12 # Esto da error porque no existe el índice 6
print(pares)

```

```python
""" Como tal se tendra en cuenta la agregación, inserción y unión de listas, estos metodos son muy útiles para manejar listas en Python. """

# para agregar un elemento al final de la lista utilizamos el método append
familia = ["Alex", "Luisa", "jorge", "Maritza"]
print(familia)

familia.append("Marta")
print(familia)

# para agregar un elemento en una posición especifica utilizamos el método insert
familia.insert(1, "Jose")
print(familia)

# para unir dos listas utilizamos el método extend
abuelos = ["herminia", "pedro"]
familia.extend(abuelos)
print(familia)

```

```python
"Aca se tendra en cuenta la eliminación de elementos en una lista, estos métodos son muy útiles para manejar listas en Python."

# Como eliminar un elemento de una lista pop(i)

# familia.pop(-1)
# print(familia)
# familia.append("jose")
print(familia)
familia.pop(3)
print(familia)
```

```python
maria = ["Ana", "Maria", "Claudia", "Luisa", "Maria"]
print(maria)

maria.append("Sofia")
print(maria)

maria.insert(2, "Valeria")
print(maria)
maria.pop(6)
print(maria)

maria.extend(familia)
print(maria)
```

```python
frutas  = ["Manzana", "Banana", "Cereza", "Naranja", "Manzana"]
print(frutas)

# si queremos eliminar la primera conciencia de la lista
frutas.remove("Manzana")
print(frutas)

# si queremos eliminar / vaciar toda la lista
frutas.clear()
print(frutas)
```

```python
# Hay una parte importante en las listas que hay que tener en cuenta, y es que las listas pueden contener otras listas, es decir listas anidadas
familia = ["Juan", "Ana", "Pedro", ["Luis", "Sofia"]]
print(familia)
print(familia[3]) # accediendo a la lista anidada
print(familia[3][1]) # accediendo al primer elemento de la lista
```

```python
# Buscar y contar 
# en las lisatas podemos buscar un elemento con el método index 

carros = ["Mazda", "Toyota", "Honda", "Chevrolet", "Ford", "Toyota"]
print(carros)

print(carros.index("Ford")) # devuelve el índice del primer elemento que coincide con el valor buscado
print("Mazda" in carros) # devuelve True si el elemento está en la lista, de lo contrario devuelve False
print(carros.count("Toyota")) # devuelve el número de veces que un elemento aparece en la lista
```

```python
# Ordenar y organizar 
nums = [10, 20, 30, 20, 11, 29, 26.2, 5, 3, 1, 100]
print(nums)
nums.sort() # ordena la lista en orden ascendente
# print(nums)

ordenada = sorted(nums) # devuelve una nueva lista ordenada sin modificar la original
print(ordenada)
print(ordenada.reverse()) # invierte el orden de la lista
print(ordenada)
```

```python
# copy en py es un metodo que permite crear una copia superficial de una lista, es decir, una nueva lista que contiene los mismos 
# elementos que la original, pero que es independiente de ella. Esto significa que si modificas la copia, la lista original 
# no se verá afectada y viceversa.

alumnos = ["Ana", "Luis", "Carlos", "Marta", "Luis", "Luisa", "Ana", "Pedro", "Sofia", "Valeria"]
print(alumnos)

copia_alumnos = alumnos.copy()
print(copia_alumnos)
copia_alumnos.remove("Ana")
print(copia_alumnos)

# Otro metodo util que es importante conocer es el list comprehension, que es una forma concisa de crear listas en Python.
# Permite generar una nueva lista aplicando una expresión a cada elemento de una secuencia o iterable, opcionalmente filtrando 
# los elementos que cumplen una condición.

cuadrados = [x**2 for x in range(16)]
print(cuadrados)
```
