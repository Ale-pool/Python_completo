 Manejo de excepciones en Python -- sin curso 
 Una exepción es un evento que interrumpe el flujo de mi codigo, cuando ocurre un error.
 Podemos manejar las excepciones usando bloques try y except.

print(1/ 0)  # Esto generará una excepción ZeroDivisionError
 aqui lo que esta haciendo python es detener la ejecucion del programa y mostrar un mensaje de error. ( division by zero )

 una expecion es la forma en la que py, te avisa de que algo salio mal en tu codigo. que ocurrio algo inesperado. el manejo de exepciones permite controlar ,
 esos errores y tomar acciones para que el programa no se detenga abruptamente


son importantes:
- permite que el script no finalice abruptamente
- ayuda a detectar, registrar y reaccionar ante errores.
- facilitan depuerar y mantener el codigo
- son clave para aplicaciones robustas, seguras y profesionales( exepcialmente en ETL, analisis de datos, APIS, ETC), FASTAPIS T DEMAS.

Extructura:

try: 
 - codigo que podria generar un error

except: 
 - que hacer si ocurre el error



```python
try:
    x = int(input("Ingrese un numero: "))
    print(f"dividir el numero ingresado {10 / x}")
except: 
    print("Se produjo un error")


print("El programa continua...")
```

```python
# captura de excepciones especificas
"""
Capturar todas las excepciones (como arriba) no es buena práctica.
Un experto identifica qué tipo de error puede ocurrir y lo maneja de forma diferente:""" 

try:
    x = int(input("Ingrese un numero: "))
    print(f"dividir el numero ingresado {10 / x}")
except ZeroDivisionError:
    print("No se puede dividir por cero")
except ValueError:
    print("Debe ingresar un numero valido")


print("El programa continua...")


# extrucura completa de manejo de excepciones
"""
try:
    # Código que puede fallar
except TipoDeError:
    # Qué hacer si ese error ocurre
else:
    # Se ejecuta solo si no hubo errores
finally:
    # Se ejecuta siempre, haya o no errores
"""
```

```python
try:
    x = int(input("Número: "))
    resultado = 10 / x
except ZeroDivisionError:
    print("Error: división por cero.")
except ValueError:
    print("Error: valor no numérico.")
else:
    print("Todo salió bien, el resultado es:", resultado)
finally:
    print("Ejecución finalizada.")
```

```python
# creacion de exepciones personalizadas

class EdadVerificarError(Exception):
  pass

def validad_edad(edad):
  if edad < 0:
      raise EdadVerificarError("La edad no puede ser negativa")
  print("Edad valida :", edad)

try:
  validad_edad(-1)
except EdadVerificarError as e:
  print("Error personalizado:",e)
```

```python
# hagamos un ejemplo con una validacion de roles

class AccesoDenegadoError(Exception):
    pass   

def validar_rol(rol):
    if rol != "admin":
        raise AccesoDenegadoError("Acceso denegado: rol no autorizado")
    print("Acceso concedido para el rol:", rol)

try:
    validar_rol("usuario")
except AccesoDenegadoError as e:
    print("Error personalizado:", e)
```

```python
# usando raise para lanzar errores

x = 10
if x < 0:
    raise ValueError("El valor no puede ser negativo ")
print("El valor es:", x)
```

```python
# capturar multiples excepciones a la vez


try:
  a = int(input("A: "))
  b = int(input("B: "))
  print( a / b)

except(ZeroDivisionError, ValueError) as e:
  print("Ocurrio un error", e)


# otro ejemplo

try:
    lista = [1, 2, 3]
    indice = int(input("Ingrese un índice: "))
    print(lista[indice])
except IndexError:
    print("Error: índice fuera de rango.")
except ValueError:
    print("Error: valor no numérico.")
except Exception as e:
    print("Ocurrió un error inesperado:", e)


```

```python
# Registrar errores sin detener el programa (logging)
# cuando estamos en proyectos grandes es mejor usar logging en lugar de print, para registrar errores, esto permite guardar los errores en un archivo de log y tener un mejor control de los mismos.
import logging

logging.basicConfig(filename='app.log', level=logging.ERROR)

def dividir(a, b):
    try:
        return a / b
    except ZeroDivisionError as e:
        logging.error("Error de división por cero: %s", e)
    except Exception as e:
        logging.error("Error inesperado: %s", e)


dividir(10, 0)

# Ahora el error se guarda en errores.log, útil para auditorías o producción.
```

```python
# ejemplo realista manejo en buvcle
while True:
    try: 
        x  = float(input("Ingrese un numero: "))
        print("Raiz cuadrada:", x ** 0.5)
        break
    except ValueError as e:
        print("Error: entrada no válida. Por favor ingrese un número.")
    except Exception as e:
        print("Ocurrió un error inesperado:", e)


```

```python
# casos en lo que es importante utilizar el finally

try:
    archivo = open("datosq.txt", "r") 
    contenido = archivo.read()
    print(contenido)
except FileNotFoundError:
    print("Error: el archivo no existe.")
finally:
    archivo.close() 
    print("El archivo ha sido cerrado.")


print("Programa finalizado.")

```

```python
class EleccionCarroError(Exception):
    pass

def eligir_carro(color):
    colores_disponibles = ["rojo", "azul", "verde"]
    if color not in colores_disponibles:
        raise EleccionCarroError(f"Color '{color}' no disponible. Elija entre {colores_disponibles}.")
    print(f"Carro de color {color} seleccionado.")

try:
    eligir_carro("azul")
except EleccionCarroError as e:
    print("Error personalizado:", e)
except TypeError as e:
    print("Error de tipo:", e)
except Exception as e:
    print(type(e).__name__, ":", e)

```
